```python
from PyPDF2 import PdfReader
import glob
import os
import pandas as pd
```

```python
def splitLine(line) :
    #print(line)
    itemIDX = line.split(' ', 1)[0]
    itemContext = line.split(' ', 1)[1].split(' Schrijf op: ')[0]
    itemTargetWord = line.split(' ', 1)[1].split(' Schrijf op: ')[1]

    return itemIDX, itemContext, itemTargetWord
```

```python
path = '/vol/bigdata3/datasets3/dutch_child_audio/letterster/spelling/cito_dictations/01_prompts/Woordenlijst_groep5_afnamekaart_E5'

file = (os.path.join(path, 'Woordenlijst_groep5_afnamekaart_E5.pdf'))
file = (os.path.join(path, 'Woordenlijst_groep6_afnamekaart_E6.pdf'))

reader = PdfReader(file)

# Page 1 (Taak 2)
page = reader.pages[0]
txt_file_page = page.extract_text().split('\n')
print(txt_file_page)
doc = txt_file_page[3:]
print(doc)

df = pd.DataFrame([splitLine(line) for line in doc], columns=['idx', 'context', 'target'])
df.to_csv(os.path.join(path, os.path.basename(file).replace('.pdf', '-taak1.tsv')), sep='\t', index=None)

# Page 2 (Taak 2)
page = reader.pages[1]
txt_file_page = page.extract_text().split('\n')
print(txt_file_page)
doc = txt_file_page[3:]
print(doc)

df = pd.DataFrame([splitLine(line) for line in doc], columns=['idx', 'context', 'target'])
df.to_csv(os.path.join(path, os.path.basename(file).replace('.pdf', '-taak2.tsv')), sep='\t', index=None)
```
